* ìŠ¬ë¼ì´ì‹±ì˜ ê¸°ë³¸ êµ¬ì¡°ëŠ” [start:stop:step]ì…ë‹ˆë‹¤.<br><br>



1. start: ìŠ¬ë¼ì´ìŠ¤ì˜ ì‹œì‘ ì¸ë±ìŠ¤ì…ë‹ˆë‹¤. ì´ ì¸ë±ìŠ¤ì˜ ìš”ì†ŒëŠ” ìŠ¬ë¼ì´ìŠ¤ì— í¬í•¨ë©ë‹ˆë‹¤.<br> startë¥¼ ìƒëµí•˜ë©´, stepì´ ì–‘ìˆ˜ì´ë©´ 0, stepì´ ìŒìˆ˜ì´ë©´ -1ë¡œ ê°„ì£¼ë©ë‹ˆë‹¤.



2. stop: ìŠ¬ë¼ì´ìŠ¤ì˜ ë ì¸ë±ìŠ¤ì…ë‹ˆë‹¤.<br> ì´ ì¸ë±ìŠ¤ì˜ ìš”ì†ŒëŠ” ìŠ¬ë¼ì´ìŠ¤ì— í¬í•¨ë˜ì§€ ì•ŠìŠµë‹ˆë‹¤.<br> stopì„ ìƒëµí•˜ë©´, stepì´ ì–‘ìˆ˜ì´ë©´ ì‹œí€€ìŠ¤ì˜ ë, stepì´ ìŒìˆ˜ì´ë©´ ì‹œí€€ìŠ¤ì˜ ì‹œì‘ìœ¼ë¡œ ê°„ì£¼ë©ë‹ˆë‹¤.
