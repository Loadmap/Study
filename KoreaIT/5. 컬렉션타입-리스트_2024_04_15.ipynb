# **7. 리스트의 함수와 메서드**

### 7-1. len(): 객체의 길이(항목의 개수)를 반환하는 내장 함수

```python
li1 = [1, 3, 5, 7, 9]

print(len(li1))
```

### 7-2. append(): 리스트에 사용되며, 리스트의 끝에 새로운 항목을 하나만 추가하는 메서드

```python
li1 = [1, 3, 5, 7, 9]

print(li1)

li1.append(0)

print(li1)

# li1.append(0, 10, 20)와 같은 경우 하나만 추가하는 메서드기에 인수가 여러개이면 append는 에러발생

li1.append([0, 10, 20])

print(li1)
```

### 7-3. extend():  리스트에 iterable(반복 가능한 객체)의 모든 항목을 추가하는 메서드

```python
li1 = [1, 3, 5, 7, 9]

print(li1)

li1.extend([0, 10, 20])

print(li1)
```

### 7-4. pop(): 리스트에서 항목을 삭제하고, 삭제된 항목을 반환하는 메서드

```python
li1 = [1, 3, 4, 5, 6]

print(li1)

print(li1.pop()) # default 는 마지막을 지우게 된다.
print(li1)

temp = li1.pop(2)
print(temp)
print(li1)
```

### 7-5. insert(): 리스트의 특정 인덱스에 항목을 추가하는 메서드

```python
li1 = [1, 3, 4, 5, 6]

print(li1)

li1.insert(1, 100)
print(li1)

li1.insert(3, [1000, 2000])
print(li1)
```

### 7-6. index(): 리스트에서 특정 값의 인덱스를 반환하는 메서드. 특정 값의 인덱스가 없다면 에러발생

```python
li1 = [1, 3, 4, 5, 6]

print(li1)

print(li1.index(3))
# print(li1.index(100)) # 해당 인덱스가 존재하지 않아 오류 발생
```

### 7-7. reverse(): 리스트의 항목들의 순서를 뒤집는 메서드


```python
li8 = [100, 50, 80, 60, 20]

print(li8)

li8.reverse() # 반대로 저장: in-place 연산

print(li8)
```

### 7-8. sort(): 리스트의 항목들을 정렬하는 메서드. inplace 연산을 수행

```python
li8 = [100, 50, 80, 60, 20]

print(li8)

li8.sort() # 오름차순, sort() default는 오름차순

print(li8)

li8.sort(reverse = True) # 내림차순

print(li8)
```

```python
li9 = ['Apple','apple', 'Grape', 'orange', 'banana', 'melon']

print(li9)

li9.sort()
print(li9)

li9.sort(reverse = True)
print(li9)
```

```python
li10 = ['사과', '바나나', '포도', '딸기']

print(li10)

li10.sort()
print(li10)

li10.sort(reverse = True)
print(li10)
```

### ✔ 비교
### sorted():  iterable(반복 가능한 객체)의 모든 항목들을 정렬한 후, 그 결과를 새로운 리스트에 담아 반환하는 함수


```python
li8 = [100, 50, 80, 60, 20]

print(li8)

print(sorted(li8)) # in-place 연산을 하지 않는다.
print(li8, '\n')

result = sorted(li8) # 오름차순. default는 오름차순
print(result)

result = sorted(li8, reverse = True) # 내림차순
print(result)
```

### 7-9. 슬라이싱을 사용하여 리스트의 순서를 뒤집는 방법

* 슬라이싱(slicing)은 파이썬에서 시퀀스 타입의 객체(예: 리스트, 문자열, 튜플 등)의 일부를 추출하는 방법입니다.
* 슬라이싱의 기본 구조는 [start:stop:step]입니다.<br><br>



1. start: 슬라이스의 시작 인덱스입니다. 이 인덱스의 요소는 슬라이스에 포함됩니다.<br> start를 생략하면, step이 양수이면 0, step이 음수이면 -1로 간주됩니다.



2. stop: 슬라이스의 끝 인덱스입니다.<br> 이 인덱스의 요소는 슬라이스에 포함되지 않습니다.<br> stop을 생략하면, step이 양수이면 시퀀스의 끝, step이 음수이면 시퀀스의 시작으로 간주됩니다.



3. step: 슬라이스의 스텝(증감) 값입니다. step을 생략하면 1로 간주됩니다.

```python
li8 = [100, 50, 80, 60, 20]

print(li8)

print(li8[0 : 5])
print(li8[0 : 5 : 1])
print(li8[ : 5 : 1])
print(li8[ : : 1])
print(li8[ : :], '\n')

print(li8[ :  : -1])
print(li8[4 : : -1])
```

### 7-10. count(): 리스트에서 특정 값의 개수를 반환하는 메서드

```python
li11 = [0, 20, 30, 50, 20, 30, 40, 20]

print(li11)

print(li11.count(20))
print(li11.count(100))
```
