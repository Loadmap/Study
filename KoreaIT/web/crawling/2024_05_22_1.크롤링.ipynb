# **1. 크롤링과 스크레이핑**

* 크롤링(crawling) : 인터넷의 데이터를 활용하기 위해 정보들을 분석하고 활용할 수 있도록 수집하는 행위

* 스크레이핑(scraping) : 크롤링 + 데이터를 추출하고 가공하는 행

# **2. Basic English Speaking**

```python
import requests
from bs4 import BeautifulSoup


```

```python
site = 'https://basicenglishspeaking.com/daily-english-conversation-topics/'

request = requests.get(site)

print(request)
# print(request.text)
```

```python
soup = BeautifulSoup(request.text)
```

```python
divs = soup.find('div', {'class' : 'thrv-columns'})

print(divs)
```

```python
links = divs.find_all('a')

print(links)
```

```python
for link in links :
    print(link.text)
```

```python
subject = []

for link in links :
    subject.append(link.text)

print(subject)
print(len(subject))
```

```python
print(f'총 {len(subject)}개의 주제를 찾았습니다.')
print()

for i in range(len(subject)) :
    print(f'{i + 1}. {subject[i]}')
```

# **3. 다음 뉴스 기사**

```python
# https://v.daum.net/v/20240522180910371
# https://v.daum.net/v/20240522195745806
# https://v.daum.net/v/20240522201402156

def daum_news_title(news_id) :
    url = f'https://v.daum.net/v/{news_id}'
    request = requests.get(url)
    soup = BeautifulSoup(request.text)
    title = soup.find('h3', {'class' : 'tit_view'})
    if title :
        return title.text.strip()

    return '제목 없음'
```

```python
daum_news_title('20240522180910371')
```

```python
daum_news_title('20240522195745806')
```

```python
daum_news_title('20240522201402156')
```

# **4. 벅스 뮤직 차트**


```python
site = 'https://music.bugs.co.kr/chart'
```

```python
# 1위. Supernova - aespa

request = requests.get(site)

soup = BeautifulSoup(request.text)

divs = soup.find('div', {'id' : 'CHARTrealtime'})

music = divs.find_all('p', {'class' : 'title'})
artist = divs.find_all('p', {'class' : 'artist'})

```

```python
list1 = []

for m in music :
    list1.append(m.text)

print(list1)
```

```python
list2 = []

for a in artist :
    list2.append(a.text.strip().split('\n'))

print(list2)
```

```python
for i, (m, a) in enumerate(zip(music, artist)) :
    print(f'{i + 1}. {m.text.strip()} - {a.text.strip()}')
```

```python
for i, (m, a) in enumerate(zip(music, artist)) :
    music = m.text.strip()
    artist = a.text.strip().split('\n')[0]

    print(f'{i + 1}위. {music} - {artist}')
```

```python
for i, (m, a) in enumerate(zip(music, artist)) :
    music = m.text.strip()
    artist = a.text.strip()

    print(f'{i + 1}위. {music} - {artist}')
```

# **6. 네이버 증권**

```python
# https://finance.naver.com/item/main.naver?code=005380
# https://finance.naver.com/item/main.naver?code=028300
# 종목 이름, 가격, 종목 코드, 거래량
# {'name' : , 'price' : , 'code' : , 'volume' : }

site = f'https://finance.naver.com/item/main.naver?code=005380'

request = requests.get(site)
soup = BeautifulSoup(request.text)

div1 = soup.find('div', {'class' : 'wrap_company'})

name = div1.text.split()[0]
code = div1.find('span', {'class' : 'code'}).text

price = soup.find('p', {'class' : 'no_today'}).text.split()[0]

volume = soup.find('table', {'class' : 'no_info'}).text.split()[12]

finance = {'종목 이름' : name, '가격' : price, '종목 코드' : code, '거래량' : volume}

print(name)
print(code)
print(price)
print(volume)
print(finance)
```

```python
def naver_finance(code) :
    site = f'https://finance.naver.com/item/main.naver?code={code}'

    request = requests.get(site)
    soup = BeautifulSoup(request.text)

    div1 = soup.find('div', {'class' : 'wrap_company'})

    name = div1.text.split()[0]
    code = div1.find('span', {'class' : 'code'}).text
    price = soup.find('p', {'class' : 'no_today'}).text.split()[0]
    volume = soup.find('table', {'class' : 'no_info'}).text.split()[12]
    # volume = table_no_info.find_all('td')[2].find('span', {'class' : 'blind'}).text

    finance = {'종목 이름' : name, '가격' : price, '종목 코드' : code, '거래량' : volume}

    return finance
```

```python
result = naver_finance('000660')

print(result)
```

```python
codes = ['000660', '005380', '028300']

data = []

for code in codes :
    dic = naver_finance(code)
    data.append(dic)

print(data)
```
