# **1. 딕셔너리**

파이썬의 딕셔너리는 키-값 쌍을 저장하는 변경 가능한(mutable) 컬렉션입니다.

```python
dic1 = {}

print(dic1)
print(type(dic1))
```

```python
dic2 = {'첫번째' : '사과', '두번째' :  '바나나', 3 : '오렌지', 4 : '포도'}

print(dic2)
print(type(dic2))
```

```python
print(dic2['첫번째'])
print(dic2[3])
```

```python
dic3 = {'num' : 1, 'userid' : 'apple', 'name' : '사과', 'tp' : '010-0021-0000'}

print(dic3)
print(dic3['num'])
print(dic3['userid'])
```

```python
dic4 = {1 : 'apple'}

print(dic4)

dic4[100] = '바나나'
print(dic4)

dic4['오렌지'] = 'orange'
print(dic4)

dic4[100] = 'banana'
print(dic4)
```

```python
del dic4[100]

print(dic4)
```

# **2. 키, 값의 제약**

딕셔너리의 키는 변경 불가능한(immutable) 타입이어야 합니다.<br> 예를 들어, 문자열, 정수, 튜플은 딕셔너리의 키로 사용할 수 있지만, 리스트는 딕셔너리의 키로 사용할 수 없습니다.<br> 하지만 딕셔너리의 값은 어떤 타입이든 상관없습니다.

```python
dic3 = {'num' : 1, 'userid' : 'apple', 'name' : '사과', 'tp' : '010-0021-0000'}

print(dic3)

dic3['num']= '10' # 변경
print(dic3)

dic3['gender'] = 'F' # 추가
print(dic3)

dic3['score'] = [90, 100, 50, 60] # 값은 리스트형도 추가가 가능
print(dic3)

# dic3[[1, 2, 3]] = ['하나', '둘', '셋']
# print(dic3) # 키 값은 변하지 않는 타입이여야 한다.

dic3[(1, 2, 3)] = ['하나', '둘', '셋'] # 튜플일 경우 변하지 않는 타입이여서 가능
print(dic3)

dic3['fruit'] = {'apple' : '🍎', 'strawberry' : '🍓', 'grape' : '🍇'}
print(dic3)
```

# **3. 딕셔너리의 메서드**


```python
dic3 = {'num' : 1, 'userid' : 'apple', 'name' : '사과', 'tp' : '010-0021-0000'}

print(len(dic3))
```

### 3-1. keys(): 딕셔너리의 모든 키를 반환하는 메서드


```python
print(dic3.keys())
print(type(dic3.keys()))
```

### 3-2. values(): 딕셔너리의 모든 값을 반환하는 메서드

```python
print(dic3.values())
print(type(dic3.values()))
```

### 3-3. items(): 딕셔너리의 모든 키-값 쌍을 반환하는 메서드

```python
print(dic3.items())
print(type(dic3.items()))
```

### 3-4. get(): 특정 키의 값을 반환. 만약 그 키가 딕셔너리에 없으면, None을 반환하는 메서드

```python
dic3 = {'num' : 1, 'userid' : 'apple', 'name' : '사과', 'tp' : '010-0021-0000'}

# print(dic3['gender']) # 에러 발생
print(dic3.get('userid'))
print(dic3.get('gender'))

print(dic3.get('gender', '성별을 알 수 없음.')) # None을 대신해서 출력하는 문장
```

### 3-5. pop(): 특정 키의 값을 제거하고 반환. 만약 그 키가 딕셔너리에 없으면, KeyError가 발생하는 메서드

```python
dic3 = {'num' : 1, 'userid' : 'apple', 'name' : '사과', 'tp' : '010-0021-0000'}

pop1 = dic3.pop('tp')
print(pop1)
print(type(pop1))
print(dic3)
```

### in : 딕셔너리에 특정 키가 있는지 확인

```python
dic3 = {'num' : 1, 'userid' : 'apple', 'name' : '사과', 'tp' : '010-0021-0000'}

print('tp' in dic3)
print('gender' in dic3)
```
