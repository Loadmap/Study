# **3. í–‰ì—´ ì—°ì‚°**

* ë„˜íŒŒì´ì—ì„œëŠ” ë‹¤ì°¨ì› ë°°ì—´ì¸ ndarrayë¥¼ ì‚¬ìš©í•˜ì—¬ í–‰ì—´ ì—°ì‚°ì„ ìˆ˜í–‰
* í–‰ì—´ ì—°ì‚°ì€ ì„ í˜• ëŒ€ìˆ˜ì™€ ê´€ë ¨ì´ ê¹Šì–´ ë°ì´í„° ê³¼í•™, ë¨¸ì‹ ëŸ¬ë‹,í†µê³„ ë“± ë‹¤ì–‘í•œ ë¶„ì•¼ì—ì„œ ì‚¬ìš©

```python
arr1 = np.array([[1, 2, 3],
                 [4, 5, 6]])

arr2 = np.array([[3, 5, 6],
                 [1, 2, 3]])

print(arr1.shape)
print(arr2.shape)
```

```python
# í–‰ì—´ ë§ì…ˆ

arr1 + arr2
```

```python
# í–‰ì—´ ëº„ì…ˆ

arr1 - arr2
```

```python
# í–‰ì—´ ì›ì†Œë³„ ê³±ì…ˆ

arr1 * arr2
```

```python
# í–‰ì—´ ë‚˜ëˆ—ì…ˆ

arr1 / arr2
```

```python
# í–‰ì—´ê³±(matrix multiplication)
# í–‰ì—´ê³±ì˜ ì¡°ê±´ : ë§ë‹¿ëŠ” shapeê°€ ê°™ì•„ì•¼í•¨. ë–¨ì–´ì ¸ ìˆëŠ” shapeê°€ ê²°ê³¼ í–‰ì—´ì´ ë¨
# (2, 3) @ (3, 2) -> (2, 2)
# ì„ í˜• ë³€í™˜ì„ í‘œí˜„í•˜ê³  ê³„ì‚°í•˜ëŠ” ë„êµ¬ë¡œ ì‚¬ìš©(í–‰ì—´ì´ ê³µê°„ì„ ë‹¤ë¥¸ ê³µê°„ìœ¼ë¡œ ë³€í™˜)
# ì„ í˜• ì—°ë¦½ ë°©ì •ì‹ì„ í–‰ì—´ í˜•íƒœë¡œ ë‚˜íƒ€ë‚´ê³  ì´ë¥¼ í’€ê¸° ìœ„í•´ í–‰ì—´ ê³±ì„ ì‚¬ìš©


arr3 = np.array([[1, 2, 3],
                 [2, 3, 4],
                 [1, 2, 3]])

arr4 = np.array([[1, 2],
                 [3, 4],
                 [5, 6]])

print(arr3.shape)
print(arr4.shape)
```

```python
print(arr3 @ arr4)
```

```python
np.dot(arr3, arr4)
```

```python
# ì „ì¹˜ í–‰ì—´
# ê¸°ì¡´ í–‰ì—´ì˜ í–‰ê³¼ ì—´ì„ ë°”ê¾¼ ìƒˆë¡œìš´ í–‰ì—´

print(arr1)
print(arr1.T)
```

```python
# ì—­í–‰ì—´
# ì£¼ì–´ì§„ ì •ì‚¬ê° í–‰ì—´ì— ëŒ€í•œ ê³±ì…ˆ ì—°ì‚°ìœ¼ë¡œ ë‹¨ìœ„ í–‰ì—´ì„ ì–»ì„ ìˆ˜ ìˆëŠ” í–‰ì—´
# ë‹¨ìœ„ í–‰ì—´: ì£¼ëŒ€ê°ì„ ì˜ ì›ì†Œê°€ ëª¨ë‘ 1ì´ê³  ë‚˜ë¨¸ì§€ ì›ì†Œê°€ ëª¨ë‘ 0ì¸ ì •ì‚¬ê°í˜• í–‰ì—´

arr = np.array([[1, 2],
                [3, 4]])

print(np.linalg.inv(arr))
print(arr @ np.linalg.inv(arr))
```
