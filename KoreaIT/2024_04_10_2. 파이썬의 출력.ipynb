# **1. print 함수**

print() 함수는 파이썬의 기본 출력 함수로, 터미널이나 콘솔 창에 텍스트나 데이터를 출력하기 위해 사용됩니다.

```python
print('Hello, World')
print(100)
print('100')
print(10, 20, 30, 40)
print(10, '이십', 30, '사십')

# 문자는 연산에 있어 불필요 하므로 '', ""를 사용함으로서 구분을 지어 놓는다.
```

```python
print(10, 20, 30, 40)

# sep: 값들 사이에 들어갈 구분자를 지정합니다. default 는 공백

print(10, 20, 30, 40, sep = '🎨', end = '\n\n')

# end: 모든 값들이 출력된 후 마지막에 출력 될 문자를 지정합니다. default 는 줄바꿈(\n)

print(10, 20, 30, 40, end = ' ')
print('python')
print('Hello, World')
```

# **2. 파이썬의 escape 문자**

* 파이썬에서 이스케이프 문자는 문자열 안에서 특별한 의미를 가지는 문자들을 나타내기 위해 사용됩니다.
* 이스케이프 문자는 백슬래시(\)로 시작하며, 문자열 내에서 특정한 동작이나 특수 문자를 표현하기 위해 사용됩니다.

```python
# \n: 줄바꿈

print ('Hello,\n World!', end = '\n\n')

# \t: 탭 문자

print('이름:\t홍길동', end = '\n\n')

# \\: 백슬래쉬 문자 표현

print('Enjoy Python! \\', end = '\n\n')

# \''or \': 큰 따옴표나 작은 따옴표를 표현

print("'안녕하세요'")
print('"안녕하세요"')
print("\"안녕하세요\"", end = '\n\n')
```

# **3.  print 함수 내의 연산**

* print() 함수는 문자열뿐만 아니라 숫자와 연산 결과도 출력할 수 있습니다.

```python
print(10)
print(10 + 5)
print('덧셈: ', 10 + 10)
print('뺄셈: ', 20 - 5)
print('곱셈: ', 5 * 20)
print('나눗셈: ', 100 / 20)
print('몫: ', 100 // 20)
print('나머지: ', 10 % 4)
print('거듭제곱: ', 10 ** 3)
```

# **4. 출력 서식 지정**

* 파이썬의 print() 함수는 출력 서식을 지정하여 값을 형식화하고 출력하는 기능을 제공합니다.
* 서식 지정자와 함께 사용되는 % 연산자와 format() 메서드를 사용하여 원하는 형식으로 출력할 수 있습니다.

```python
print('정수: %d' %100)
print('정수: %d' %100.5)
```

```python
print('실수: %f' %10.234)
print('실수: %0.2f' %11.456)
```

```python
print('문자열: %s' %100)
print('문자열: %s' %'pyhton')
```

```python
# f-string

n = 100

print(f'정수: {n}')
```

```python
print('5 + 3 = 8')
print('%d + %d = %d, \n' %(5, 3, 5 + 3))

a = 5
b = 3
c = a + b

print(f'{a} + {b} = {c}')
```

```python
print('이름: %s, 나이: %d' %('홍길동', 30))
print('몸무게: %f' %50.1234)
print('몸무게: %0.2f' %50.1234)
print('내 친구: {}, {}, \n'.format('오렌지', '바나나'))

a = '오렌지'
b = '바나나'

print(f'내 친구: {a}, {b}')
```

# **5. 주석문**

* 파이썬 주석은 코드 내에 프로그래머가 작성한 설명이나 메모를 포함하는데 사용됩니다.
* 주석은 코드를 이해하고 다른 개발자와 협업하는 데 도움이 되며, 코드의 기능과 목적을 명확하게 설명할 수 있습니다.
* 주석은 프로그램 실행 중에 무시되므로 코드 실행에 영향을 미치지 않습니다.

```python
# 문자열 출력은 작은 따옴표 또는 큰 따옴표를 사용 합니다.

print('파이썬 출력하기')
```

```python
'''
여러줄을 작성하는주석: 큰 따옴표 또는 작은 따옴표 3개 사이에 작성합니다. 여러줄을 작성하는데 있어서 편리합니다.
'''

print('여러줄 주석 작성하기!')
```
