# **1. matplotlib**

* 파이썬 기반 시각화 라이브러리
* 한글에 대한 지원이 아직 완벽하지 않음
* pandas 와 연동이 용이함
* [Matplotlib 공식 홈페이지](https://matplotlib.org)

```python
# matlab과 비슷하게 명령어 스타일로 동작하는 함수들의 모음

import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1, 2, 3, 4]) # 리스트의 값들은 y값들이며 x값들은 자동으로 생성

plt.show()
```

```python
plt.plot(['one', 'two', 'three', 'four'], [1, 2, 3, 4])

plt.show()
```

```python
data = np.arange(1, 100)

plt.plot(data)

plt.show()
```

```python
data1 = np.arange(1, 50)

plt.plot(data1)

data2 = np.arange(50, 100)

plt.plot(data2)

plt.show()
```

```python
# 여러개의 plot를 그리는 방법
# subplot(행 개수, 열 개수, plot번호)

data1= np.arange(1, 50)
plt.subplot(2, 1, 1)
plt.plot(data1)

data2 = np.arange(50, 100)
plt.subplot(2, 1, 2)
plt.plot(data2)
```

```python
plt.rc('font', family = 'NanumBarunGothic')
```

```python
plt.figure(figsize = (3, 4)) # inch 단위 사이즈 조정

plt.plot([1, 2, 3], [1, 2, 3])

plt.title('제목', fontsize = 30)
plt.xlabel('x축', fontsize = 20)
plt.ylabel('y축', fontsize = 20, rotation = 0)

plt.show()
```

```python
plt.figure(figsize  = (13, 10))

plt.plot(np.arange(10), np.arange(10), color = 'deepskyblue', marker = 'o', linestyle = '-.', ms = 10)
plt.plot(np.arange(10), np.arange(10) * 2, color = 'deeppink', linestyle = '--', marker = 'x')
plt.plot(np.arange(10), np.arange(10) * 3, color = 'gold', marker = 'v')

plt.title('마커설정', fontsize = 30)

# 범례
# 기본 위치 : 왼쪽 상단

plt.legend(['10', '10 * 2', '10 * 3'], fontsize = 15, loc = 'lower right', ncol = 3)

plt.xlim(0, 12)
plt.ylim(0, 30)

plt.xticks(rotation = 30)
plt.yticks(rotation = 30)

plt.show()
```

```python
x = ['파이썬', '데이터 분석', '웹개발', '머신러닝', '딥러닝', '컴퓨터비전']
y = [95, 70, 80, 50, 30, 32]

plt.figure(figsize = (8, 5))

plt.title('AI 성적표', fontsize = 25)
plt.ylabel('수강생 점수', fontsize = 10)

plt.bar(x, y, alpha = 0.5, color = 'deepskyblue')

plt.show()
```

```python
x = ['파이썬', '데이터 분석', '웹개발', '머신러닝', '딥러닝', '컴퓨터비전']
y = [95, 70, 80, 50, 30, 32]

plt.figure(figsize = (8, 5))

plt.title('AI 성적표', fontsize = 25)
plt.xlabel('수강생 점수')

plt.barh(x, y, alpha = 0.5, color = 'purple')

plt.show()
```
