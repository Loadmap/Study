# **1. Iris DataSet**

* 데이터셋 : 특정한 작업을 위해 데이터를 관련성있게 모아놓은 것
* [사이킷런 데이터셋 페이지](https://scikit-learn.org/stable/api/sklearn.datasets.html#module-sklearn.datasets)

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

# iris
```

```python
print(iris['DESCR'])
```

```python
data = iris['data']

data[:5]
```

```python
target = iris['target']

target
```

```python
feature_names = iris['feature_names']

feature_names
```

* sepal length (cm) : 꽃받침의 길이
* sepal width (cm) : 꽃받침의 너비
* petal length (cm) : 꽃잎의 길이
* petal width (cm) : 꽃잎의 너비

```python
import pandas as pd
```

```python
df_iris = pd.DataFrame(data, columns = feature_names)

df_iris.head()
```

```python
target.shape
```

```python
df_iris['target'] = target

df_iris.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# train_test_split(독립변수, 종속변수, 테스트사이즈, 시드값...)

X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target', axis = 1),
                                                    df_iris['target'],
                                                    test_size = 0.2,
                                                    random_state = 2024)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)

y_pred
```

```python
print('정답률 : ', accuracy_score(y_test, y_pred))
```

```python
# 정답률이 만족스럽지 않기 때문에 다시 데이터를 섞고자 함

X_train, X_test, y_train, y_test = train_test_split(df_iris.drop('target', axis = 1),
                                                    df_iris['target'],
                                                    test_size = 0.2,
                                                    random_state = 2023)
```

```python
X_train.head()
```

```python
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)

print('정답률 : ', accuracy_score(y_test, y_pred))
```

```python
# 6.7, 3.3, 5.7, 2.1

y_pred = svc.predict([[6.8, 3.2, 5.6, 2.2]])

y_pred
```
