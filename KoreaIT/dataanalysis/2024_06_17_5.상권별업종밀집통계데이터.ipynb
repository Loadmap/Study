* [folium 공식 홈페이지](https://python-visualization.github.io/folium)

```python

# !pip install folium
```

```python
import folium
```

```python
# foliu.Map(location = [위도, 경도], zoom_start = 배열...)

map_folium = folium.Map(location = [37.500043085219545, 127.03557801473566], zoom_start = 17)

map_folium
```

```python
# Marker() : location 속성으로 지정한 위치에 마커를 생성
# Popup() : 마커를 클릭했을때 표시할 문자열을 설정
# add_to() : 작성된 마커를 지도에 추가

map_folium = folium.Map(location = [37.500043085219545, 127.03557801473566], zoom_start = 17)
popup = folium.Popup('코리아IT아카데미', max_width = 200)
folium.Marker(location = [37.500043085219545, 127.03557801473566], popup = popup).add_to(map_folium)


map_folium
```

```python
def markerset(name, location, map, icon, color) :
    popup = folium.Popup(name, max_width = 200)
    folium.Marker(location = location, popup = popup,
                  icon = folium.Icon(icon = icon, color = color),).add_to(map)
```

```python
from folium.features  import CustomIcon

icon = CustomIcon('police.jpg', icon_size= (45, 45))

map_folium = folium.Map(location = [37.56652479270633, 126.9779129116262], zoom_start = 14)

markerset('서울특별시청', [37.56652479270633, 126.9779129116262], map_folium, 'star', 'red')
markerset('덕수궁', [37.56585320879458, 126.97514572197248], map_folium, 'ok', 'pink')

popup = folium.Popup('서울지방경찰청', max_width = 200)
folium.Marker(location = [37.57504045678939, 126.97195576938117], popup = popup,
                  icon = folium.Icon(icon = 'info-sign', color = 'green')).add_to(map_folium)

markerset('광화문', [37.57617987638928, 126.97694467807803], map_folium, 'star', 'orange')

# CircleMarker() : 원의 범위를 마커로 설정

popup = folium.Popup('영풍문고-종로본점', max_width = 200)
folium.CircleMarker(location = [37.56999947440024, 126.98228391170936], popup = popup,
                  color = 'purple', fill = True, fill_color = 'white').add_to(map_folium)

map_folium
```

### 문제

* shop_seoul_edu_computer 데이터프레임에 있는 모든 데이터를 지도에 표기
* 단 일반 마커로 표기하며 클릭하면 상호명-도로명주소 형태의 팝업을 보여줌

```python
data = shop_seoul_edu_com

edu_map = folium.Map(location = [data['위도'].mean(), data['경도'].mean()], zoom_start = 12)

for i in data.index :
    edu_name = data.loc[i, '상호명'] + '-' + data.loc[i, '도로명주소']

    popup = folium.Popup(edu_name, max_width = 500)
    folium.Marker(location = [data.loc[i, '위도'], data.loc[i, '경도']], popup = popup).add_to(edu_map)

edu_map
```
