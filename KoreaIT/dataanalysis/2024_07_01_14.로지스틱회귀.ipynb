# **2. 로지스틱 회귀(Logistic Regression)**

* 둘 중의 하나를 결정하는 문제(이진 분류)를 풀기 위한 대표적인 알고리즘
* 다항 분류 문제에도 확장될 수 있음
* 종속 변수 y는 두가지 범주 중 하나른 가짐 (예 : 0 또는 1)
* 특정 범주의 속할 확률을 예측하는 것

<img src = 'https://velog.velcdn.com/images/zlddp723/post/ce2bc937-d8dd-4d4f-8c22-e7baf8dcd8ce/image.png' >

로지스틱 회귀 모델은 일반화 선형 모델의 일종으로 독립 변수의 선형 조합을 로지스틱 함수(시그모이드 함수)를 사용하여 종속 변수에 대한 확률 점수로 변환

````python
from sklearn.linear_model import LogisticRegression
````

````python
lr = LogisticRegression()
````

````python
lr.fit(X_train, y_train)
````

````python
pred = lr.predict(X_test)
````

````python
from sklearn.metrics import accuracy_score
````

````python
accuracy_score(y_test, pred)
````

# **3. 혼돈 행렬(confusion metrics)**

* 정밀도 와 재현율(민감도)를 활용하는 평가용 지수

````python
from sklearn.metrics import confusion_matrix
````

````python
confusion_matrix(y_test, pred)
````

```
TN(8784)  FP(100)

FN(673)   TP(175)
```

* TN(진음성) : 승진을 못했지만 승진하지 못했다고 예측해서 맞춤
* FP(위양성) : 승진을 못했는데 승진했다고 예측해서 틀림
* FN(위음성) : 승진을 했는데 승진을 못했다고 예측해서 틀림
* TP(진양성) : 승진을 했는데 승진을 했다고 예측해서 맞춤

### 3-1. 정밀도(Precision)

* TP/ (TP + FP)
* 무조건 양성으로 판단해서 계산하는 방법
* 승진을 예측한 것 중에서 실제로 승진한 비율


### 3-2. 재현율(recall)

* TP / (TP + FN)
* 정확하게 감지한 양성 샘풀의 비율
* 실제 승진한 사람 중에서 승진으로 정확하게 예측한 비율
* 민감도 또는 TPR(True Positive Rate) 라고 함

### 3-3. F1 score

* 정밀도와 재현율의 조화평균을 나타내는 지표

```
정밀도 재현율 산술평균 조화평균
0.4    0.6    0.5      0.48
0.3    0.7    0.5      0.42
0.5    0.5    0.5      0.5
```
$$2*\frac{정밀도 * 재현율}{정밀도 + 재현율}=\frac{TP}{TP+\frac{FN+FP}{2}}$$

````python
from sklearn.metrics import precision_score, recall_score, f1_score
````

````python
precision_score(y_test, pred)
````

````python
recall_score(y_test, pred)
````

````python
f1_score(y_test, pred)
````

# **4. 교차 검증(Cross Validation)**

* train_test_split 에서 발생하는 데이터의 섞임에 따라 성능이 좌우되는 문제를 해결하기 위한 방법
*  k겹(k-flod) 교차 검증을 가장 많이 사용

````python
from sklearn.model_selection import KFold
````

````python
kf =  KFold(n_splits = 5)

kf
````

````python
hr_df.head()
````

````python
for train_index, test_index in kf.split(range(len(hr_df))) :
    print(train_index, test_index, len(train_index), len(test_index))
````

````python
kf =  KFold(n_splits = 5, random_state = 2024, shuffle = True)

kf
````

````python
for train_index, test_index in kf.split(range(len(hr_df))) :
    print(train_index, test_index, len(train_index), len(test_index))
````

````python
acc_list = []

for train_index, test_index in kf.split(range(len(hr_df))) :
    X = hr_df.drop('is_promoted', axis = 1)
    y = hr_df['is_promoted']

    X_train = X.iloc[train_index]
    X_test = X.iloc[test_index]
    y_train = y.iloc[train_index]
    y_test = y.iloc[test_index]

    lr = LogisticRegression()
    lr.fit(X_train, y_train)
    pred = lr.predict(X_test)
    acc_list.append(accuracy_score(y_test, pred))
````

````python
acc_list
````

````python
np.array(acc_list).mean()
````

교차 검증을 사용하는 이유는 결과를 좋게 하기 위함이 아니라 믿을만한 검증을 하기 위함
