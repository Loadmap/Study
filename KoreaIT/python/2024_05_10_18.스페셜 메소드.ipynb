# **1. 스페셜 메서드**

파이썬의 스페셜 메서드 (또는 매직 메서드라고도 불림)는 더블 언더스코어(__)로 시작하고 끝나는 메서드 이름을 갖습니다. 이 메서드들은 특정 구문이나 내장 함수를 사용할 때 파이썬 인터프리터에 의해 자동으로 호출됩니다.

예를 들어, 객체에 대해 + 연산자를 사용하면 해당 객체의 \_\_add__ 메서드가 호출됩니다. 또는 len() 함수를 사용하면 \_\_len__ 메서드가 호출됩니다.

```python
class Book :

    def __init__(self, title) :
        self.title = title
```

```python
book = Book('재밌는 파이썬')

print(book)
print(str(book))
```

```python
class Book :

    def __init__(self, title) :
        self.title = title

    def __str__(self) :
        return self.title
```

```python
book = Book('재밌는 파이썬')

print(book)
print(str(book))
print(Book.mro())
```

```python
li = [1, 2, 3, 4, 5]

print(li)
print(str(li))
```

```python
class Calc :

    def __init__(self, num) :
        self.num = num

    def __add__(self, other) :
        return self.num + other.num
```

```python
num1 = Calc(4)
num2 = Calc(2)

print(num1 + num2)
```

```python
class Queue :

    def __init__(self) :
        self.items = [1, 2, 3, 4, 5]

    def __len__(self) :
        return len(self.items)
```

```python
li = [1, 2, 3, 4, 5]

print(len(li))
```

```python
queue = Queue()

print(queue)
print(len(queue))
```

```python
class Point :

    def __init__(self, x, y) :
        self.x = x
        self.y = y

    def __getitem__(self, index) :
        if index == 0 :
            return self.x

        elif index == 1 :
            return self.y

        else :
            return -1
```

```python
pt = Point(5, 3)

print(pt[0])
print(pt[1])
print(pt[2])
```
