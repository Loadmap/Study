# **1. 반복문**

반복문은 동일한 작업을 여러 번 실행하기 위해 사용되는 제어 구조입니다. 주로 for 문과 while 문이 사용되며, 각각의 반복문은 다른 상황에 적합한 방식으로 사용됩니다.

# **2. while 문**

while 문은 특정 조건이 참인 동안 반복적으로 코드 블록을 실행하는 제어 구조입니다. while 문은 주어진 조건이 참인 동안 반복적으로 코드를 실행하며, 조건이 거짓이 되면 반복을 멈춥니다.

```
while 조건:
    # 조건이 참인 동안 실행될 코드
```

````python
i = 1

while i <= 5 :
    print(f'Hello, python {i}')
    i += 1
````

````python
# 1 부터 10 까지의 총합

i = 1
sum = 0

while i <= 10 :
    sum += i
    i += 1

print(f'1부터 10 까지의 합 : {sum}')
````

### 문제1

1부터 100까지의 짝수의 합을 구하는 프로그램을 작성해보

# **3. for 문**

for 문은 시퀀스(리스트, 튜플, 문자열 등)의 각 항목에 대해 반복 작업을 수행하는 반복문입니다. for 문은 주로 "지정된 범위 내에서 반복"할 때 사용됩니다. 이것은 반복할 요소들이 미리 정해져 있는 경우에 유용합니다.

```
for 요소 in 시퀀스:
    # 요소에 대한 작업 수행
```

````python
list1 = [10, 20, 30, 40]

for li in list1 :
    print(li)
````

### 3-1. range()

range() 함수는 파이썬에서 순차적인 정수의 시퀀스를 생성하는 데 사용되는 내장 함수입니다.<br> 주로 for 문과 함께 사용되며, 일정한 범위 내에서 반복 작업을 수행할 때 유용하게 활용됩니다.<br> range() 함수의 결과로 생성되는 순차적인 정수 시퀀스는 실제로 메모리에 저장되지 않고 필요할 때마다 생성되므로, 큰 범위의 반복문에도 효율적으로 사용할 수 있습니다.

```
range([start], stop, [step])
```

 * start (선택 사항): 순차적인 범위의 시작 값을 지정합니다. 기본값은 0입니다.
 * stop: 순차적인 범위의 끝 값을 지정합니다. 생성된 시퀀스는 stop 값 직전까지의 정수를 포함합니다.
 * step (선택 사항): 순차적인 값을 증가시키는 간격을 지정합니다. 기본값은 1입니다.

````python
num = range(1, 11, 1)

print(num)
print(type(num))
````

````python
for i in num :
    print(i, end = ' ')
````

````python
# step은 생략 가능하고 default 는 1
# start는 생략 가능하고 default 는 0

for i in range(1, 11):
    print(i, end = ' ')

print('')

for j in range(11) :
    print(j, end = ' ')
````

````python
for i in range(1, 101, 2) :
        print(i, end = ' ')
````

### 3-1. enumerate()

enumerate() 함수는 파이썬에서 반복문을 사용할 때 인덱스와 값을 함께 가져오기 위해 사용되는 내장 함수입니다.<br>
주로 for 문과 함께 사용되며, 반복문 내에서 순서를 추적하면서 값과 해당 값의 인덱스를 함께 사용할 때 유용합니다.<br>
enumerate() 함수는 (인덱스, 값)의 튜플을 반환하는 이터레이터(iterator)를 생성합니다.

```
enumerate(iterable, start=0)
```

* iterable: 반복 가능한 객체(리스트, 튜플, 문자열 등)를 지정합니다.
* start (선택 사항): 인덱스의 시작 값을 지정합니다. 기본값은 0입니다.
enumerate() 함수는 (인덱스, 값) 형태의 튜플을 생성하여 이터레이터를 반환합니다. 이를 for 문과 함께 사용하여 인덱스와 값을 동시에 가져와 처리할 수 있습니다.

````python
list1 = [10, 20, 30, 40]

for i in enumerate(list1, 0) :
    print(i)
````

````python
list1 = [10, 20, 30, 40]

for i in enumerate(list1, 10) :
    print(i)
````

````python
list1 = [10, 20, 30, 40]

for i in enumerate(list1) :
    print(i)
````

````python
list1 = [10, 20, 30, 40]

# (0, 10)
# i, v =(0, 10)

for i, j in enumerate(list1, 0) :
    print(f'index : {i}, value : {j}')
````

### 3-3. zip()

zip() 함수는 파이썬에서 여러 개의 반복 가능한(iterable) 객체를 병렬적으로 묶어주는 내장 함수입니다.<br>
각 객체로부터 순서대로 하나의 요소씩 가져와 튜플로 묶어주는 이터레이터를 반환합니다.<br>
주로 여러 개의 리스트나 튜플을 병렬적으로 처리하고자 할 때 사용됩니다.

```
zip(iterable1, iterable2, ...)
```

````python
# li1의 요소가 변경되면 for문의 코드가 에러를 일으키기에 for 문도 수정해야 한다.

li1 = [10, 20, 30]
li2 = ['apple', 'banana', 'orange']

for i in range(0, 3):
    print(li1[i], li2[i])
````

````python
li1 = [10, 20]
li2 = ['apple', 'banana', 'orange']

if len(li1) == len(li2) :
    for i in range(len(li1)):
        print(li1[i], li2[i])

elif len(li1) > len(li2) :
    print(f'li1 요소 : {len(li1)} 개, li2 요소 : {len(li2)} 개로 요소 개수가 일치 하지 않습니다.')

else :
    for i in range(len(li1)):
        print(li1[i], li2[i])
````

````python
li1 = [10, 20, 30]
li2 = ['apple', 'banana', 'orange']

for i, j in zip(li1,li2) :
    print(i, j)
````

````python
# 에러가 발생하지 않도록 맞춰서 실행한다.

li1 = [10, 20, 30, 40]
li2 = ['apple', 'banana', 'orange']

for i in zip(li1,li2) :
    print(i)
````

````python
for n, A, a in zip('12345', 'ABCDE', 'abcde') :
    print(n, A, a)
````

# **4. 중첩 반복문**

파이썬에서 중첩된 반복문은 하나의 반복문 안에 다른 반복문이 포함된 구조를 의미합니다.<br> 중첩 반복문은 외부 반복문의 각 반복마다 내부 반복문이 모두 실행되는 방식으로 작동합니다.<br> 이를 통해 다차원의 데이터를 처리하거나 복잡한 작업을 수행할 수 있습니다.

```
for 외부_변수 in 외부_시퀀스:
    for 내부_변수 in 내부_시퀀스:
        # 내부 반복문 코드
    # 외부 반복문 코드
```

* 외부_변수: 외부 반복문의 현재 요소를 나타내는 변수입니다.
* 외부_시퀀스: 외부 반복문에서 순회할 반복 가능한 객체(리스트, 튜플 등)입니다.
* 내부_변수: 내부 반복문의 현재 요소를 나타내는 변수입니다.
* 내부_시퀀스: 내부 반복문에서 순회할 반복 가능한 객체입니다.

````python
start1 = 1
end1 = 4

start2 = 1
end2 = 4

for i in range(start1, end1) :
    print(f'외부 cnt : {i} 바퀴')
    print('')

    for j in range(start1, end1) :
        print(f'\t 내부 cnt : {j} 바퀴')

    print('')

print('')
print(f'총 바퀴수 : {i * j} 바퀴')
````

### 문제4

2단부터 9단까지 구구단을 출력하는 프로그램을 만들어 보자

- 단 for 문과 while 문을 사용하여 별도로 작

````python
for i in range(2, 10) :
    print(f'{i} 단')
    print()
    for j in range(1, 10) :
        print(f'{i} * {j} = {i * j}')
    print()
````

````python
i = 2
j = 1

while i <= 9 :
    print(f'{i}단')
    print()

    while j <= 9 :
        print(f'{i} * {j} = {i * j}')
        j += 1

    i += 1
    j = 1
    print()
````
