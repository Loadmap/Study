# **1. 사용자 정의 함수**

사용자 정의 함수란 사용자가 특정 작업을 수행하기 위해 직접 작성한 함수를 의미합니다.<br> 파이썬에는 많은 내장 함수들이 있지만, 때로는 우리의 요구사항에 맞게 동작하는 함수를 직접 만들어야 할 때가 있습니다.<br> 이때 사용자 정의 함수를 작성하게 됩니다.<br> 기본적인 구조는 아래와 같습니다.

```
def 함수명(매개변수1, 매개변수2, ...):
    # 함수 내용
    return 결과값
```

### 1-1. 매개변수와 반환값이 없는 함

````python
def func1() :
    print('처음으로 만드는 함수')
````

````python
func1()
func1()
func1()
````

````python
for i in range(5) :
    func1()
````

### 1-2. 매개변수가 있고 반환 값이 없는 함수

````python
def func2(num) :
    print(f'입력받은 숫자 : {num}')
````

````python
# func2() # 사용자 함수 func2 에 num에 대입되는 수가 없으므로 에러 발생
func2(100)
````

````python
def func3(a, b) :
    sum = 0

    for i in range(a, b + 1) :
        sum += i

    print(f'{a} 부터 {b} 까지의 합 : {sum}')
````

````python
func3(1, 100)
````

````python
def func4() :
    print('안녕하세요. 파이썬')
````

````python
func4()

temp = func4()
print(temp) # func4() 함수에 반환 값이 없기에 None 출력
````

````python
temp = func4

print(f'temp에 저장된 값 : {temp}')
````

````python
temp()
````

### 1-4. 기본값이 설정된 매개변수

````python
def func6(num1 =  0, num2 = 0) : # 매개변수 num1, num2의 default 를 0 으로 설정
    sum = num1 + num2

    return sum ;
````

````python
print(func6())
print(func6(10))
print(func6(10, 3))
print(func6(num2 = 3))
````

### 1-5. 가변 매개변수

함수를 호출할 때 *를 사용하면 시퀀스(리스트, 튜플 등)의 요소를 개별적인 위치 인자로 풀어서 전달할 수 있습니다.

````python
def func7(*args) :
    return (args)
````

````python
print(func7())
print(func7(10, 30))
print(func7(10, 30, 70, 100))
````

### 1-6. 키워드 매개변수



````python
def func8(id, name, age) :
    print(f'id : {id}')
    print(f'name : {name}')
    print(f'age : {age}')
````

````python
func8('apple', '사과', 20)
````

````python
func8(age  = 20, id = 'apple', name = '사과')
````

````python
dic1 = {'age' : 20, 'id' : 'apple', 'name' : '사과'}
func8(**dic1) # ** 시 dictionary 형태의 key, value를 매개변수에 전달할때 사용
````

````python
func8(*dic1) # * 시에는 dictionary 형태의 key 값만 매개변수에 전달
````

### 1-7. 여러개의 반환값

````python
def func9(num1, num2) :
    return num1 + num2
````

````python
result1 = func9(10, 3)

print(result1)
````

````python
def func10(num1, num2) :
    return num1 + num2, num1 - num2, num1 * num2, num1 / num2
````

````python
result1, result2, result3, result4 = (func10(10, 5))

print(f'덧셈 : {result1}, 뺄셈 : {result2}, 곱셈 : {result3}, 나눗셈 : {result4}')
````

````python
result1, result2, result3, result4 = (func10(10, 5))

print(f'곱셈 : {result3}') # 나머지 3 변수도 메모리에 저장되어있다
````

````python
_, _, result3, _ = (func10(10, 5))

print(f'곱셈 : {result3}') # _ 는 해당 변수는 사용하지 않겠다는 의미
````

````python
li1 = [10, 20, 30, 40, 50]

for i, v in enumerate(li1) :
    print(f'값 : {v}')
````

# **2. None 의 특징**

None은 파이썬에서 특별한 값으로, 아무런 값이 없음을 표현하는 데 사용됩니다.<br> 다른 언어에서의 null 또는 nil과 유사한 개념입니다.

None은 파이썬의 내장 상수이며, 그 자체로 데이터 타입이 NoneType입니다.<br> 모든 None은 동일하므로, 두 개의 None 값을 비교할 때 항상 True를 반환합니다.

````python
num = None

print(num)
````

````python
def func11(num) :

    if num >= 10 :
        return num
````

````python
print(func11(10))
print(func11(4))
````

````python
def func12(num = None) :

    if num is None :
        print('전달된 값이 없습니다.')

    else :
        print(num)
````

````python
func12(10)
func12()
````

````python
num = None

if num is None :
    print('num의 값은 없습니다.')

# == 를 통해 None을 비교하는 것은 권장하지 않음

if num == None :
    print('num의 값은 없습니다.')
````

````python
li1 = [1, 2, 3, 4, 5]
li2 = [1, 2, 3, 4, 5]

# 객체 비교

result1= li1 is li2

print(result1)

# 값 비교

result2 = li1 == li2

print(result2)
````
