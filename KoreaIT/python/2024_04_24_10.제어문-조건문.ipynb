# **1. 조건문**

조건문은 코드의 실행 흐름을 제어하기 위해 사용하는 구문입니다. 이 조건문은 주어진 조건이 참인지 거짓인지에 따라 다른 코드를 실행하게 합니다.

# **2. if 문**

가장 기본적인 조건문입니다. 특정 조건이 참(True)일 경우에만 해당 블록 내의 코드가 실행됩니다.





```
if 조건:
    실행할 코드
```

````python
bool1 = True

if bool1 :
    print('결과는 참입니다.')

print('\n프로그램을 종료 합니다.')
````

````python
# str1 = 'Hello'
str1 = ''

if str1 :
    print('결과는 참입니다.')

print('\n프로그램을 종료합니다.')
````

````python
age = int(input('나이를 입력하세요. : '))

if age > 19 :
    print('\n성인입니다.')
    print('넷플릭스 성인 영화를 감상 할 수 있습니다.')

print(f'\n입력한나이 : {age}살')
````

# **3. else 문**

else는 파이썬의 조건문 구조에서 if와 elif 조건들이 모두 거짓으로 평가될 경우 실행되는 부분을 정의하는 데 사용됩니다.<br> else는 선택적으로 사용될 수 있으며, 그 자체로는 어떤 조건을 갖지 않습니다.<br> 다시 말해, else 블록은 위의 if와 elif 조건들이 모두 만족되지 않는 경우에만 실행됩니다.

```
if 조건1:
    코드 블록1
elif 조걸2:
    코드 블록2
else:
    코드 블록3
```




````python
age = int(input('나이를 입력하세요. : '))

if age > 19 :
    print('\n성인입니다.')
    print('넷플릭스 성인 영화를 감상 할 수 있습니다.')

else :
    print('\n미성년자 입니다.')
    print('디즈니 플러스를 보세요.')

print(f'\n입력한나이 : {age}살')
````

# **4. elif 문**

elif는 if-else 구조에서 추가적인 조건을 검사하기 위해 사용되는 키워드입니다.<br> elif는 "else if"의 줄임말이며, 주어진 여러 조건 중에서 하나만 참일 때 실행되는 코드 블록을 정의합니다.

```
if 조건1:
    코드 블록1
elif 조건2:
    코드 블록2
elif 조건3:
    코드 블록3
else:
    코드 블록4
```

````python
age = int(input('나이를 입력하세요. : '))

if age > 19 :
    print('\n성인입니다.')

elif age > 14 :
    print('\n청소년 입니다.')

elif age > 6 :
    print('\n어린이 입니다.')

elif age > 0 :
    print('\n유아 입니다.')

else :
    print('0 초과되는 숫자를 입력해주세요.')

print(f'\n입력한나이 : {age}살')
````

````python
kor, eng, math = input('국어,영어,수학점수를 입력하세요').split(' ')
tot = int(kor) + int(eng) + int(math)
avg = tot / 3
hak = 'F학점'
if avg >= 90:
    hak = 'A학점'
elif avg >= 80:
    hak = 'B학점'
elif avg >= 70:
    hak = 'C학점'
elif avg >= 60:
    hak = 'D학점'
print('총점: ', tot)
print('평균: ', avg)
print('학점: ', hak)









````

````python
score = a, b, c = input('[국어, 영어, 수학]의 형식에 맞춰서 점수를 차례대로 입력해주세요. : ').split(', ')

dic1 = {'국어' : int(a), '영어' : int(b), '수학' : int(c)}
tot1 = dic1['국어'] + dic1['영어'] + dic1['수학']
avg1 = (dic1['국어'] + dic1['영어'] + dic1['수학']) / len(dic1)

print(dic1, '\n')

if avg1 >= 90 :
    print(f'총점 : {tot1}, 평균 {avg1}점으로 A학점 입니다.')

elif avg1 >= 80 :
    print(f'총점 : {tot1}, 평균 {avg1}점으로 B학점 입니다.')

elif avg1 >= 70 :
    print(f'총점 : {tot1}, 평균 {avg1}점으로 C학점 입니다.')

elif avg1 >= 60 :
    print(f'총점 : {tot1}, 평균 {avg1}점으로 D학점 입니다.')

else :
    print(f'총점 : {tot1}, 평균 {avg1}점으로 F학점 입니다.')
````

# **5. 조건부 표현식**

조건부 표현식은 파이썬에서 삼항연산자라고도 불립니다.<br> 간단한 if-else 구문을 더 간결하게 표현할 수 있는 방법을 제공합니다.

```
조건이 참인 경우 값 if condition else 조건이 거짓인 경우 값
```

````python
num = int(input('숫자를 입력하시오. : '))

if num % 2 == 0 :
    print('짝수')

else :
    print('홀수')
````

````python
num = int(input('숫자를 입력하시오. : '))

print('짝수') if num % 2 == 0 else print('홀수')
````

````python
age = int(input('나이를 입력하세요. : '))

if age > 19 :
    print('\n성인입니다.')

elif age > 14 :
    print('\n청소년 입니다.')

elif age > 6 :
    print('\n어린이 입니다.')

elif age > 0 :
    print('\n유아 입니다.')

else :
    print('입력을 확인하세요.')
````

````python
age = int(input('나이를 입력하세요. : '))

print('\n성인입니다.') if age > 19 else print('\n청소년 입니다.') if age > 14 else print('\n어린이 입니다.') if age > 6 else print('\n유아 입니다.') if age > 0 else print('입력을 확인하세요.')
````
