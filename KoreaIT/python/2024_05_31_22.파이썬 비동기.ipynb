# **1. 동기(Synchronous)**

* 작업을 순차적으로 실행하는 것
* 하나의 작업이 끝나기 전에는 다음 작업을 시작하지 않음

```python
# 메인 함수는 func1 함수가 종료될때까지 기다림
# 하지만 func1 함수가 종료되기 전에 main 함수에서 처리해야 할 일이 있다면?

def func1() :
    print('1')
    print('2')
    print('3')

def main() :
    func1()
```

```python
main()
```

# **2. 비동기(Asynchronous)**

* 작업이 병렬적으로 실행하는 것
* 시간이 오래 걸리는 작업을 기다리는 동안 다른 작업을 수행

```python
import time

def smile() :
    time.sleep(1)
    print('😁😁😁😁😁')

def main() :
    smile()
    smile()
```

```python
print(f'start : {time.strftime("%X")}')
main()
print(f'end : {time.strftime("%X")}')
```

✔ RuntimeError: asyncio.run() cannot be called from a running event loop

* asyncio.run() : 함수는 새로운 이벤트 루프를 생성하고 주어진 서브루틴(코루틴)을 실행하기 위해 사용
* 코렙이나 주피터 노트북 환경에서는 이미 이벤트 루프가 실행 중이기 때문에  asyncio.run()을 호출하면 에러 발생
* 따라서 아래와 같이 모듈을 import 하고 메서드를 호출 해야 함

```python
import asyncio
import nest_asyncio
nest_asyncio.apply()

async def smile() :
    await asyncio.sleep(1)
    print('😁😁😁😁😁')

async def main() :
    await asyncio.gather(
        smile(),
        smile()
    )
```

```python
print(f'start : {time.strftime("%X")}')
asyncio.run(main())
print(f'end : {time.strftime("%X")}')
```
