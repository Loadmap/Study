```python
import asyncio
import nest_asyncio
nest_asyncio.apply()

async def smile() :
    await asyncio.sleep(1)
    print('ğŸ˜ğŸ˜ğŸ˜ğŸ˜ğŸ˜')

async def main() :
    await asyncio.gather(
        smile(),
        smile()
    )
```
