# **세트**

세트(set)는 중복되지 않는 항목들의 컬렉션입니다.

```python
s1 = {1, 3, 4, 5}

print(s1)
print(type(s1))
```

```python
# data 가 없을 시 에는 dict 타입으로 인식한다.

s2 = {}

print(s2)
print(type(s2))
```

```python
s3 = set([1, 2, 3, 4, 6])

print(s3)
print(type(s3))
```

```python
li1 = [1, 2, 3, 4]
print(type(li1), '\n')

s4 = set(li1)

print(s4)
print(type(s4))
```

```python
s5 = {1, 3, 4, 5, 5, 6, 7, 8, 1, 6}

print(s5)
```

```python
li2 = [1, 3, 4, 5, 5, 6, 7, 8, 1, 6]
print(li2, '\n')

s6 = set(li2)
print(s6)
```

# **2. 세트의 메서드**



### 2-1. add() : 세트의 요소를 추가

```python
s1 = {1, 3, 4, 5}

s1.add(2)
print(s1)

s1.add(6)
print(s1)

s1.add(3)
print(s1)
```

### 2-2. update(): 세트에 여러 항목을 한번에 추가하는 메서드

```python
s1 = {1, 3, 5, 7}

s1.update([2, 4, 6, 8, 10])
print(s1)
```

### 2-3. remove():  세트의 항목을 제거. 만약 그 항목이 세트에 없으면, KeyError가 발생

```python
s1 = {1, 3, 5, 7}

s1.remove(3)
print(s1)

# s1.remove(4)
# print(s1) # 항목이 세트에 없으면, KeyError가 발생
```

### 2-4. discard(): 세트의 항목을 제거. 제거할 항목이 없어도 에러가 발생하지 않음

```python
s1 = {1, 3, 5, 7}

s1.discard(3)
print(s1)

s1.discard(4)
print(s1)
```

### 2-5. copy(): 세트를 복사

```python
s1 = {1, 3, 5, 7}

s2 = s1

print(s1)
print(s2, '\n')

print(id(s1))
print(id(s2), '\n')

s_1 = s1.copy()
print(s_1)
print(id(s_1))
```

### 2-6. union(): 두 세트의 합집합을 반환

```python
s1 = {10, 20, 30, 40, 50}
s2 = {30, 40, 50, 60, 100}

result1 = s1.union(s2)
print(result1)
```

```python
result2 = s1 | s2

print(result2)
```

### 2-7. intersection(): 두 세트의 교집합을 반환

```python
s1 = {10, 20, 30, 40, 50}
s2 = {30, 40, 50, 60, 100}

result1 = s1.intersection(s2)
print(result1)
```

```python
result2 = s1 & s2

print(result2)
```

### 2-8. difference(): 두 세트의 차집합을 반환

```python
s1 = {10, 20, 30, 40, 50}
s2 = {30, 40, 50, 60, 100}

result1 = s1.difference(s2)
print(result1)
```

```python
result2 = s1 - s2

print(result2)
```

### 2-9. symmetric_difference(): 대칭 차집합을 계산

```python
s1 = {10, 20, 30, 40, 50}
s2 = {30, 40, 50, 60, 100}

result1 = s1.symmetric_difference(s2)
print(result1)
```

```python
result2 = s1 ^ s2
print(result2)
```
