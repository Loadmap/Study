# **1. 튜플**

여러 가지 데이터 타입의 값을 저장할 수 있는, 변경 불가능(immutable)한 순서가 있는 컬렉션입니다.

```python
tu1 = (1, 2, 3, 4, 5)

print(tu1)
print(type(tu1))
```

```python
tu2 = (10) # tu2 = 10

print(tu2)
print(type(tu2), '\n')

# 요소가 1개일 경우에는 끝에 , 를 꼭 붙여야한다.

tu2 = (tu2,)
print(tu2)
print(type(tu2))
```

```python
tu3 = tuple([1, 3, 4, 5, 6])

print(tu3)
print(type(tu3))
```

```python
tu4 = 1, 3, 5, 6

print(tu4)
print(type(tu4))
```

```python
tu5 = ('apple', 'banana', ('🍎', '🍌'))

print(tu5)
print(type(tu5), '\n')

print(tu5[0])
print(type(tu5[0]), '\n')

print(tu5[2][1])
print(type(tu5[2][1]), '\n')

print(tu5[2])
print(type(tu5[2]))
```

# **2. input() 함수**

* input() 함수는 사용자로부터 키보드 입력을 받는 데 사용되는 내장 함수입니다.
* 프로그램 실행 중에 사용자가 값을 입력하면, 입력한 값이 문자열로 반환되어 변수에 저장하거나 처리할 수 있습니다.

```python
name = input(f'이름을 입력하시오 : ')
```

```python
print(name)
```

```python
print(f'{name}님 안녕하세요!')
```

```python
# 사과, 20살, 50Kg

name, age, weight = input('이름, 나이, 몸무게를 , 기준으로 입력하세요. :').split(', ')

print(f'{name}님의 나이는 {age}살 몸무게 {weight}Kg 입니다.')
```

```python
# int() : 문자열을 정수 타입으로 변환

num1 = input('첫번째 숫자를 입력하세요: ')
num2 = input('두번째 숫자를 입력하세요: ')

print('결과: ', int(num1) + int(num2))
```

### 문제

* 국어점수, 영어점수, 수학점수를 입력받아 총점, 평균을 구하는 프로그램을 작성해보자

* 단, 입력방법은 input 함수를 이용하여 100/70/ 40 형태로 입

```python
li = a, b, c = input('국어점수/ 영어점수/ 수학점수 를 앞의 형식과 같이 입력해주세요. : ').split('/ ')
print(li)
print(len(li),'\n')

sum = int(a) + int(b) + int(c)
avg = (sum / len(li))

print(f'국어, 영어, 수학의 총점은 {sum}, 평균은 {avg : .2f} 입니다.')
```
