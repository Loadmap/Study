# **1. 스코프**

파이썬에서 변수의 범위(scope)는 해당 변수가 프로그램 내에서 참조되고 변경될 수 있는 영역을 의미합니다.<br> 파이썬의 변수 범위는 크게 네 가지로 분류됩니다

### 1-1. 지역변수(Local)

 변수가 함수 내에서 정의될 때 해당 함수 내부에서만 사용함을 의미합니다.

```python
def local_ex():
    local_var = '로컬 변수'
    print(local_var)
```

```python
local_ex()

# print(local_var) # 에러 발생 로컬 변수이기 때문에 loca_ex 안에서만 사용이 가능하다.
```

### 1-2. 둘러싼 범위(Enclosing)

Enclosing은 내부 함수에서 외부 함수의 변수를 참조 할 수 있다.

```python
def outer_func() :
    enclosing_var = '둘러싼 범위 변수'

    def inner_func() :
        print(enclosing_var)

    inner_func()
```

```python
outer_func()
```

### 1-3. Global(전역 범위)

Global 범위는 변수가 최상위 수준에서 정의될 때 전체 파일 내에서 해당 변수를 의미합니다.

```python
global_var = '글로벌 변수'

def test_global() :
    print(global_var)
```

```python
test_global()
```

```python
# 전역변수와 지역변수

global_var = 10 # 전역변수

def test_global() :
    global_var = 20 # 지역변수
    print(global_var)

test_global()
print(global_var)
```

```python
global_var = 10 # 전역변수

def test_global() :
    global global_var
    global_var = 20 # 전역변수를 수정
    print(global_var)

test_global()
print(global_var)
```

### 1-4. Built-in(내장 범위)

Built-in scope는 파이썬의 내장 함수와 모듈들이 속하는 범위입니다. 예를 들어, print(), len() 같은 함수들이 여기에 해당됩니다.

```python
print(print)
print(len)
```

# **2. 변수의 범위 탐색 순서**

* 1. Local scope
* 2. Enclosing scope
* 3. Global scope
* 4. Built-in scope

따라서 지역 범위에 동일한 이름의 변수가 없으면 파이썬은 둘러싼 범위를 확인하고, 그 다음으로 전역 범위, 마지막으로 내장 범위를 확인합니다.<br>

이렇게 변수의 범위를 이해하고 관리하는 것은 코드의 가독성과 유지 보수성을 높이고, 예기치 않은 오류를 방지하는 데 중요합니다.
