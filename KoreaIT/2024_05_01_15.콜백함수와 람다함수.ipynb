# **1. 콜백 함수**

콜백 함수는 다른 함수에 인자로 전달되어, 어떤 이벤트나 특정 조건에서 실행되게 되는 함수를 말합니다.<br> 콜백 함수는 주로 비동기 작업이나 특정 이벤트 발생 시점에 어떤 동작을 수행할 때 사용됩니다.

```python
def callback_func(func) :
    for i in range(5) :
        func()
```

```python
def print_hello() :
    print('안녕하세요! 파이썬!')
```

```python
print_hello()
```

```python
callback_func(print_hello)
```

```python
def callback_func(func, num) :
    for i in range(num) :
        func(i)
```

```python
def print_hello(num) :
    print('안녕하세요! 파이썬!', num)
```

```python
def print_hi(num) :
    print('하이! 파이썬!', num)
```

```python
print_hello(1)
print_hi(1)
```

```python
callback_func(print_hello, 3)
```

```python
callback_func(print_hi, 5)
```

# **2. 람다 함수**

람다 함수는 파이썬에서 익명의 간단한 함수를 생성하기 위한 특별한 구문입니다.<br> "익명의 함수"라는 것은 함수에 고유한 이름이 지정되지 않았음을 의미합니다.<br> 람다 함수는 일반적인 함수(def를 사용하여 정의)와는 달리, 한 줄로 표현되는 짧고 간결한 함수를 생성할 때 주로 사용됩니다.

```python
def square(x) :
    return x ** 2
```

```python
print(square(5))
```

```python
square = lambda x : x ** 2
```

```python
print(square(5))
```

```python
(lambda x : x ** 2)(5)
```

```python
people = [
    {'name' : '오렌지', 'age' : 30},
    {'name' : '사과', 'age' : 20},
    {'name' : '바나나', 'age' : 35},
    {'name' : '포도', 'age' : 40}
]
```

```python
def sort_age(x) :
    return x['age']
```

```python
print(sort_age(people[0]))
```

```python
sorted_people = sorted(people, key = sort_age) # key 라는 속성은 sorted를 콜백 함수를 기준으로 조정한다는 의미
print(sorted_people)
```

```python
sorted_people = sorted(people, key = lambda x : x['age']) # labmda 함수는 무조건 return형
print(sorted_people)
```

# **3. 람다가 유용하게 사용되는 대표적인 함수**

### 3-1. filter 함수

filter()는 파이썬의 내장 함수로, 주어진 함수의 조건을 만족하는 항목만으로 이루어진 이터레이터를 반환합니다.<br> 이 함수는 주로 리스트나 다른 순차적인 데이터 타입에서 특정 조건을 만족하는 항목들만을 필터링할 때 사용됩니다.

```python
li = [2, 5, 7, 10, 15, 17, 20, 22, 25, 28]

def even(n) :
    if n % 2 == 0 :

        return True

    else :

        return False
```

```python
result = list(filter(even, li))

print(result)
```

```python
result2 = list(filter(lambda x : x % 2 == 0, li))

print(result2)
```

### 3-2. map 함수

map()는 파이썬의 내장 함수로, 주어진 함수를 이터러블의 모든 항목에 적용하여 결과를 반환하는 이터레이터를 생성합니다.<br> 이 함수는 주로 리스트나 다른 순차적인 데이터 타입의 항목 각각에 함수를 적용할 때 사용됩니다.

```python
numbers = [1, 2, 3, 4, 5]

squared_num = list(map(lambda x : x ** 2, numbers))

print(squared_num)
```

```python
li1 = [1, 2, 3]
li2 = [4, 5, 6]

sum = list(map(lambda x, y : x + y, li1, li2))

print(sum)
```

```python
words = ['apple', 'banana', 'cherry']

upper_words= list(map(lambda x : x.upper(), words))

print(upper_words)
```
