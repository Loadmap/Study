### 이스케이프 코드

프로그래밍을 할 때 사용할수 있도록 미리 정의해둔 '문자조합'

* \n:	문자열 안에서 줄을 바꿀 때 사용
* \t:	문자열 사이에 탭 간격을 줄 때 사용
* \\:	\를 그대로 표현할 때 사용
* \':	작은따옴표(')를 그대로 표현할 때 사용
* \":	큰따옴표(")를 그대로 표현할 때 사용
* \r:	캐리지 리턴(줄 바꿈 문자, 커서를 현재 줄의 가장    앞으로 이동)
* \f:	폼 피드(줄 바꿈 문자, 커서를 현재 줄의 다음 줄로 이동)
* \a:	벨 소리(출력할 때 PC 스피커에서 '삑' 소리가 난다)
* \b:	백 스페이스
* \000:	널 문자

```python
print('Hello world \n')
print('Hello\tworld')
print('Hello world \\')
print('Hello world \'')
print('Hello world \"')
print('Hello world \r')
print('Hello world \f')
print('Hello world \a')
print('Hello world \b')
print('Hello world \000')
```

### 문자열 인덱싱과 슬라이싱

```python
a = 'It\'s a nice morning for a bike ride.'

print(f'a[3] = {a[3]}')
print(f'a[5] = {a[5]}')
print(f'a[7] = {a[7]}')

# 인덱싱(Indexing)은 프로그래밍에서 배열, 리스트, 문자열 등과 같은 데이터 구조에서
# 각 요소에 번호를 부여하여 해당 요소를 식별하는 과정이나 방법을 말합니다.
# 첫글자 를 시작으로 0부터 숫자를 셉니다.
```

```python
# 첫번째 문자를 기준으로 뒤에서 부터 n번째 문자를 가르킬때

print(f'a[-3] = {a[-3]}')
print(f'a[-5] = {a[-5]}')
print(f'a[-7] = {a[-7]}')
```

```python
# 단순히 문자열에서 한 문자만 뽑아내는것이 아니라 한 단어를 뽑아내는 방법
# 슬라이싱을 이용한다.

a = 'It\'s a nice morning for a bike ride.'

print(f'a[7:11] = {a[7:11]}')

# 슬라이싱은 a[시작번호 : 끝번호]로 지정되어지고 끝번호에 해당되는 번호의 문자는 포함되어지지 않는다.

print(f'a[:] = {a[:]}')

# 슬라이싱에서 시작번호와 끝번호를 생략할 시 해당 문자열을 전부 출력한다

print(f'a[7:-11] = {a[7:-11]}')

# 슬라이싱도 인덱싱과 마찬가지로 -기호를 사용 할 수 있다.
```

```python
# 슬라이싱으로 문자열 나누기

a = '20230918sunny'

date = a[:8]
weather = a[8:]

print(date)
print(weather)
```

```python
year = a[:4]
month = a[4:6]
day = a[6:8]

print(year)
print(month)
print(day)
```

### 문자열 포맷팅

문자열 포맷 코드

* %s:	문자열(String)
* %c:	문자 1개(character)
* %d:	정수(Integer)
* %f:	부동소수(floating-point)
* %o:	8진수
* %x:	16진수
* %%:	Literal % (문자 % 자체)

```python
a = 2
b = 3

print('I want to eat {num1} apples and {num2} bananas'.format(num1 = a, num2 = b))
```

```python
# 왼쪽, 오른쪽, 가운데 정렬
hi = "hi"

print('{0:<10}'.format('hi'))
print(f'{hi:<10}')

print('{0:>10}'.format('hi'))
print(f'{hi:>10}')

print('{0:^10}'.format('hi'))
print(f'{hi:^10}')
```

```python
# 공백채우기

hi = "hi"

print('{0:=<10}'.format('hi'))
print(f'{hi:!^10}')
```

```python
# 소수점 표현하기

a = 3.141592

print('{0:0.4f}'.format(a))
print(f'{a:0.4f}')
```

```python
# {} 와 같은 중괄호 문자를 포매팅문자가 아닌 문자 그래도 사용하고 싶은 경우

a = 3.141592

print('{{a}}'.format())
print(f'{{{a:.3f}}}')
```

### 문자열 관련 함수들

```python
# 문자 개수 세기 - count

a = 'playground'

print(a.count(a))

# 문자 위치 알려주기 1 - find

b = 'I want to play with my friend'

print(b.find('a'))

# 문자 위치 알려주기 2 - index

print(b.index('a'))

# 문자열 삽입 - join

print('!'.join('??'))
print(','.join(['a', 'b', 'c', 'd']))
```

```python
# 소문자를 대문자로 - upper

a = 'hi'

print(a.upper())

# 대문자를 소문자로 - lower

b = 'HI'

print(b.lower())

# 왼쪽 공백 지우기 - lstrip

c = '   hi   '

print(c.lstrip())

# 오른쪽 공백 지우기 - rstrip

print(c.rstrip())

# 양쪽 공백 지우기 - strip

print(c.strip())
```

```python
# 문자열 바꾸기 - replace

a = 'I can\'t do anything'

print(a.replace('can\'t', 'can'))

# 문자열 나누기 - split

print(a.split())

b = 'I,can,do,anything'

print(b.split(','))
```

```python
# index 를 이용한 인덱스 반환

a = [1, 2, 3]

print(a.index(3))

# index(x)의 x의 값에 해당하는 인덱스 값을 출력한다 ex) a.index(3)는 3의 값이
# 리스트 a의 3번째에 해당하는 위치에 있으므로 2를 출력한다.
```

```python
# insert 를 이용한 리스트에 요소 삽입

a = [1, 2, 3]

a.insert(0, 4)

print(a)

# insert(위치, 삽입할 요소)

a.insert(2, 5)

print(a)
```

```python
# remove 를 이용한 리스트 요소 제거

a = [1, 2, 3, 2, 3, 1]

a.remove(2)

print(a)

# remove는 리스트에서 첫번째로 나오는 요소 x를 삭제한다.
```

```python
# pop 을 이용한 리스트 요소 끄집어내기

a = [1, 2, 3]

print(a.pop())
print(a)

# pop 은 해당 리스트 x번째 요소를 리턴하고 그 요소를 리스트에서 삭제한다.

b = [1, 2, 3, 4]

print(b.pop(2))
print(b)
```

```python
# count 를 이용한 리스트의 포함된 요소 x의 갯수 세기

a = [1, 2, 3, 2, 1, 2, 4]

print(a.count(2))
```

```python
# extend를 이용한 리스트 확장

a = [1, 2 , 3]

a.extend([3, 4, 5])

print(a)

# a += [3, 4, 5]와 동일한 코드라 볼 수 있다.
```

## 딕셔너리 자료형

딕셔너리는 key 와 value 값을 한쌍으로 가지는 자료형이다.

### 딕셔너리 쌍추가, 삭제하기

```python
# 쌍 추가하기

a = {1 : 'a'}

a[2] = 'b'

print(a)

# 딕셔너리 요소 삭제하기

del a[2]

print(a)
```

### 딕셔너리에서 key를 사용해서 value 얻기

```python
student = {'name' : 'James', 'age' : 16, 'Phonenumber' : '023-423-4556'}

print(student['name'])
print(student['age'])
print(student['Phonenumber'])

# 단 key 가 중복되었을때는 1개를 제외한 나머지 key value 값은 모두 무시되어진다.
# 한 딕셔너리에서 key값이 중복이 될 수 없기 때문
```

### key, value 리스트 만들기

```python
student = {'name' : 'James', 'age' : 16, 'Phonenumber' : '023-423-4556'}

print(student.keys())
print(student.values())
print(list(student.keys()))
print(list(student.values()))
```

```python
# items 를 이용한 key, value 값 얻기

student = {'name' : 'James', 'age' : 16, 'Phonenumber' : '023-423-4556'}

print(student.items())
print(list(student.items()))
```

```python
# clear를 이용한 key, value 값 모두 지우기

student = {'name' : 'James', 'age' : 16, 'Phonenumber' : '023-423-4556'}

print(student.clear())
```

```python
# get 을 이용한  key 로 value 값 얻기

student = {'name' : 'James', 'age' : 16, 'Phonenumber' : '023-423-4556'}

print(student.get('name'))
print(student.get('age'))
```

```python
# in 을 이용한 해당 key 가 딕셔너리 안에 있는지 조사하기

student = {'name' : 'James', 'age' : 16, 'Phonenumber' : '023-423-4556'}

print('name' in student)
print('weight' in student)
```

## 집합 자료형

집합 자료형은 아래와 같이 2가지 특징을 지니는 자료형이다

* 중복을 허용하지 않는다.
* 순서가 존재하지 않는다.

```python
a = set([1, 2, 3, 4, 2])

print(a)

b = set('Hello')

print(b)

# 집합 자료형은 중복과 순서가 존재하지 않는다. 따라서 인덱싱을 통해 요소값을 얻을 수 없다.
# 만약 set 자료형에 저장된 값을 인덱싱으로 접근하려면 다음과 같이 리스트나 튜플로 변환 후에 해야한다.

a = list(a)
b = tuple(b)

print(a[0])
print(b[2])
```

### 교집합, 합집합, 차집합 구하기

```python
a = set([1, 2, 3, 4, 5])
b = set([3, 4, 5, 6, 9])

# 교집합

print(a & b)
print(a.intersection(b))

# 합집합

print(a | b)
print(a.union(b))

# 차집합

print(a - b)
print(a.difference(b))
```

### 집합 자료형 관련 함수

```python
# add 를 이용한 값 1개 추가하기

a = set([1, 2, 3])

a.add(5)

print(a)

# update 를 이용한 값 여러개 추가하기

b = set([1, 3, 4, 5])

b.update([2, 6, 7])

print(b)

# remove 를 이용한 특정 값 제거하기

a.remove(2)

print(a)
```

## 불 자료형

참과 거짓을 나타내는 자료형

```python
print(1 == 1)
print(1 == 2)
```

```python
a = 'python'
b = ''
c = [1, 2, 3]
d = []
e = {'a' : 1}
f = 0
g = None

list = [a, b, c, d, e, f, g]

print(list)

for i in list :
    print(i)
    if i :
        print('참')
    else :
        print('거짓')
```

```python
# bool 을 이용한 참, 거짓 판별

print(bool('python'))
print(bool(''))
```

## 자료형의 값을 저장하는 공간, 변수

```python
# 변수 메모리 주소 확인

a = [1, 2, 3]

print(id(a))
```

```python
# 리스트를 복사하고자 할때

a = [1, 2, 3]

b = a

print(b)
print(id(a))
print(id(b))

# copy 모듈 이용하기

c = a.copy()

print(c)
print(id(c))

# copy 를 사용할 시 변수의 값은 같지만 객체가 다른 것을 확인할 수 있다.
```
