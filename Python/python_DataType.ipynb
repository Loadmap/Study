## 문자열 자료형

### 이스케이프 코드

프로그래밍을 할 때 사용할수 있도록 미리 정의해둔 '문자조합'

* \n:	문자열 안에서 줄을 바꿀 때 사용
* \t:	문자열 사이에 탭 간격을 줄 때 사용
* \\:	\를 그대로 표현할 때 사용
* \':	작은따옴표(')를 그대로 표현할 때 사용
* \":	큰따옴표(")를 그대로 표현할 때 사용
* \r:	캐리지 리턴(줄 바꿈 문자, 커서를 현재 줄의 가장    앞으로 이동)
* \f:	폼 피드(줄 바꿈 문자, 커서를 현재 줄의 다음 줄로 이동)
* \a:	벨 소리(출력할 때 PC 스피커에서 '삑' 소리가 난다)
* \b:	백 스페이스
* \000:	널 문자

```python
print('Hello world \n')
print('Hello\tworld')
print('Hello world \\')
print('Hello world \'')
print('Hello world \"')
print('Hello world \r')
print('Hello world \f')
print('Hello world \a')
print('Hello world \b')
print('Hello world \000')
```

### 문자열 인덱싱과 슬라이싱

```python
a = 'It\'s a nice morning for a bike ride.'

print(f'a[3] = {a[3]}')
print(f'a[5] = {a[5]}')
print(f'a[7] = {a[7]}')

# 인덱싱(Indexing)은 프로그래밍에서 배열, 리스트, 문자열 등과 같은 데이터 구조에서
# 각 요소에 번호를 부여하여 해당 요소를 식별하는 과정이나 방법을 말합니다.
# 첫글자 를 시작으로 0부터 숫자를 셉니다.
```

```python
# 첫번째 문자를 기준으로 뒤에서 부터 n번째 문자를 가르킬때

print(f'a[-3] = {a[-3]}')
print(f'a[-5] = {a[-5]}')
print(f'a[-7] = {a[-7]}')
```

```python
# 단순히 문자열에서 한 문자만 뽑아내는것이 아니라 한 단어를 뽑아내는 방법
# 슬라이싱을 이용한다.

a = 'It\'s a nice morning for a bike ride.'

print(f'a[7:11] = {a[7:11]}')

# 슬라이싱은 a[시작번호 : 끝번호]로 지정되어지고 끝번호에 해당되는 번호의 문자는 포함되어지지 않는다.

print(f'a[:] = {a[:]}')

# 슬라이싱에서 시작번호와 끝번호를 생략할 시 해당 문자열을 전부 출력한다

print(f'a[7:-11] = {a[7:-11]}')

# 슬라이싱도 인덱싱과 마찬가지로 -기호를 사용 할 수 있다.
```

```python
# 슬라이싱으로 문자열 나누기

a = '20230918sunny'

date = a[:8]
weather = a[8:]

print(date)
print(weather)
```

```python
year = a[:4]
month = a[4:6]
day = a[6:8]

print(year)
print(month)
print(day)
```

### 문자열 포맷팅

문자열 포맷 코드

* %s:	문자열(String)
* %c:	문자 1개(character)
* %d:	정수(Integer)
* %f:	부동소수(floating-point)
* %o:	8진수
* %x:	16진수
* %%:	Literal % (문자 % 자체)

```python
a = 2
b = 3

print('I want to eat {num1} apples and {num2} bananas'.format(num1 = a, num2 = b))
```

```python
# 왼쪽, 오른쪽, 가운데 정렬
hi = "hi"

print('{0:<10}'.format('hi'))
print(f'{hi:<10}')

print('{0:>10}'.format('hi'))
print(f'{hi:>10}')

print('{0:^10}'.format('hi'))
print(f'{hi:^10}')
```

```python
# 공백채우기

hi = "hi"

print('{0:=<10}'.format('hi'))
print(f'{hi:!^10}')
```

```python
# 소수점 표현하기

a = 3.141592

print('{0:0.4f}'.format(a))
print(f'{a:0.4f}')
```

```python
# {} 와 같은 중괄호 문자를 포매팅문자가 아닌 문자 그래도 사용하고 싶은 경우

a = 3.141592

print('{{a}}'.format())
print(f'{{{a:.3f}}}')
```

### 문자열 관련 함수들

```python
# 문자 개수 세기 - count

a = 'playground'

print(a.count(a))

# 문자 위치 알려주기 1 - find

b = 'I want to play with my friend'

print(b.find('a'))

# 문자 위치 알려주기 2 - index

print(b.index('a'))

# 문자열 삽입 - join

print('!'.join('??'))
print(','.join(['a', 'b', 'c', 'd']))
```

```python
# 소문자를 대문자로 - upper

a = 'hi'

print(a.upper())

# 대문자를 소문자로 - lower

b = 'HI'

print(b.lower())

# 왼쪽 공백 지우기 - lstrip

c = '   hi   '

print(c.lstrip())

# 오른쪽 공백 지우기 - rstrip

print(c.rstrip())

# 양쪽 공백 지우기 - strip

print(c.strip())
```

```python
# 문자열 바꾸기 - replace

a = 'I can\'t do anything'

print(a.replace('can\'t', 'can'))

# 문자열 나누기 - split

print(a.split())

b = 'I,can,do,anything'

print(b.split(','))
```
