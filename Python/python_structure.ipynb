## if 문

```
기본구조

if 조건문 :
    수행할 문장1
    수행할 문장2
    ...
else :
    수행할 문장A
    수행할 문장B
```

해당 조건에 맞는 상황을 수행하는 '조건문' 이다.

````python
# 간단한 예시

# 2000원 보다 많으면 택시를 탈 수 있고 그렇지 않다면 걸어가야한다.

while True:
    money = input("Enter the amount of money (or 'q' to quit): ")

    if money.lower() == 'q':
        break

    if money.isdigit():
        money = int(money)

        if money > 2000:
            print('You can ride a taxi!')
            break

        else:
            print("You can't ride a taxi, so you walk home on foot")
            break
    else:
        print('You have to enter a valid number')
````

### and or not

* x and y :	x와 y 모두 참이어야 참이다.
* not x :	x가 거짓이면 참이다.

## while 문

```
while 조건문 :
    수행할 문장1
    수행할 문장2
    수행할 문장3
    ...
```

문장을 반복해서 수행해야 할 경우 사용하는 '반복문'이다.

````python
# 간단한 예시
import time

t = 10

while t > 0 :
    print(f'남은 시간이 {t} 초 남았습니다.')
    t -= 1
    time.sleep(1)

    if t == 0 :
        print('남은 시간이 종료되었습니다.')
````

````python
prompt = '''
1. Add
2. del
3. list
4. quit

enter number : '''

number = 0

while number != 4 :
    print(prompt)
    number = int(input())
````

### while 문 강제로 빠져나가기

while 문은 조건문이 참인 동안 계속해서 while 문 안의 내용을 반복적으로 수행한다.

때문에 while 문에서 강제적으로 빠져나가려고 할 때에 break 문을 사용해서 빠져 나올 수 있다.

````python
# 예시

coffee = 10
money = True

while money :
    print('돈이 있으므로 커피를 제공합니다.')

    coffee -= 1

    print(f'남은 커피는 {coffee}잔 입니다.\n')

    time.sleep(1)

    if coffee == 0 :
        print('남은 커피가 다 떨어졌습니다. 판매를 중지합니다.')
        break
````

### while 문의 맨 처음으로 돌아가기

while 문 안의 문장을 수행할때 입력조건을 검사해서 조건에 맞지 않아서 맨 처음 조건문으로 다시 돌아가게 만들고 싶은 경우 continue문을 사용한다.

````python
# 이를 이용해서 조금 더 복잡한 코드 작성이 가능하다.
coffee = 10

while True :

    order = input('커피를 주문하시겠습니까? (Y / N) : ')

    if order.lower() == 'y' :
        count = input(f'몇잔을 주문하시겠습니까? 최대 주문 가능량 {coffee}잔: ')

        if count.isdigit() == True :
            count = int(count)

            if count > 10 :
                print('남은 커피 수량보다 많은 값을 입력했습니다. 다시 입력해주세요.')
                continue

            price =  count * 300
            money = input(f'커피를 구매하시려면 돈을 넣어 주십시오. (필요 최소금액 {(price)}원) : ')


            if money.isdigit() == True :
                money = int(money)

                if money == price :
                    print('커피를 제공합니다.')

                    coffee = coffee - count
                    print(f'남은 커피는 {coffee} 잔 입니다')

                elif money > price :
                    print('커피를 제공합니다')
                    print(f'거스름돈이 남아 거스름돈 {money - price}원을 반환합니다.')

                    coffee = coffee - count
                    print(f'남은 커피는 {coffee} 잔 입니다.')

                else :
                    print('돈이 부족합니다. 다시 돈을 넣어주세요')
                    continue

            else :
                print('올바르게 숫자를 다시 입력해주세요.')
                continue

        else :
            print('올바르게 숫자를 다시 입력해주세요.')
            continue

    elif order.lower() == 'n' :
        print('이용해주셔서 감사합니다.')
        break

    else :
        print('잘못된 입력입니다. 다시 입력해주세요.')
        continue
````

## for문

```
for 변수 in 리스트(또는 튜플, 문자열) :
    수행할 문장1
    수행할 문장2
    ...
```

while 문 와 비슷한 반복문으로 반복이 정해진 횟수나 시퀀스에서 각 요소를 처리해야 할때 사용한다.

````python
# 예시

number = [1, 2, 3, 4, 5]

for n in number :
    print(n)
````

````python
l = ['car', 'red', 'lemon', 'bus', 'orange', 'car', 'green', 'taxi']
fruit = ['apple', 'banana', 'lemon', 'orange', 'grape']
vehicle = ['car', 'bus', 'taxi', 'airplane', 'boat']

for i in l :

    if i in fruit :
        print(f'{i}는 과일이다!')

    elif i in vehicle :
        print(f'{i}는 탈것이다!')

    else :
        print(f'{i}는 색깔이다!')
````

````python
# for 문 역시 continue, break 가 사용 가능하다.

score = [20, 70, 40, 30]
number = 0

for i in score :
    number += 1

    if i < 35 :
        continue

    print(f'{number} 번째 학생 {i} 점으로 합격입니다.')
````

````python
score = [20, 70, 40, 30]
number = 0

for i in score :
    number += 1
    print(f'{number} 번째 학생 {i} 점 입니다.')

    if max(score) == i :
        print(f'{number} 번째 학생 {i} 점으로 최고 점수 입니다.')
        break
````

````python
# for 문 과 함께 자주 사용하는 range 함수
# range(시작 숫자, 끝 숫자) 로 시작숫자 이상 끝 숫자 미만인 객체를 만든다.

a = range(10)

print(a)
````

````python
number = 0

for i in range(11) :
    number += i

print(number)
````

````python
# 이를 이용해 같은 출력값을 가지지만 다른 코드

score = [20, 70, 40, 30]

for i in range(len(score)) :

    if score[i] < 35 :
        continue

    print(f'{i + 1} 번째 학생 {score[i]} 점으로 합격입니다.')
````

````python
 # range 를 이용한 구구단

for i in range(1, 11) :

    for j in range(1, 11) :

        print(i * j, end = ' ')
    print('')
````

### 리스트 컴프리헨션 사용하기

````python
a = [1, 2, 3, 4, 5]
l = []

for n in a :
    l.append(n * 3)

print(l)

# 위 코드를 리스트 컴프리헨션을 사용하면

l = [n * 3 for n in a]
print(l)

# 같은 결과를 내면서 좀 더 간단하게 작성할 수 있다.
````

````python
# 만약 위 구구단 처럼 for문을 2번 사용하는 경우도 아래와 같이 간단하게 작성 가능하다.

r = [x * y for x in range(1, 6) for y in range(1, 6)]

print(r)
````
