## if 문

```기본구조

if 조건문 :
    수행할 문장1
    수행할 문장2
    ...
else :
    수행할 문장A
    수행할 문장B
```

````python
# 간단한 예시

# 2000원 보다 많으면 택시를 탈 수 있고 그렇지 않다면 걸어가야한다.

while True:
    money = input("Enter the amount of money (or 'q' to quit): ")

    if money.lower() == 'q':
        break

    if money.isdigit():
        money = int(money)

        if money > 2000:
            print('You can ride a taxi!')
            break

        else:
            print("You can't ride a taxi, so you walk home on foot")
            break
    else:
        print('You have to enter a valid number')
````

### and or not

* x or y :	x와 y 둘 중 하나만 참이어도 참이다.
* x and y :	x와 y 모두 참이어야 참이다.
* not x :	x가 거짓이면 참이다.

````python
# 위 코드에서 or 을 추가한 코드
# 2000원 보다 많고 카드가 있다면 택시를 탈 수 있고 그렇지 않다면 걸어가야한다.

card = True

while True:
    money = input("Enter the amount of money (or 'q' to quit): ")

    if money.lower() == 'q':
        break

    if money.isdigit():
        money = int(money)

        if money > 2000 :
            print('You can ride a taxi!')
            break

        else:
            print("You can't ride a taxi, so you walk home on foot")
            break

    elif money.lower() == 'card' :
        print('You can ride a taxi!')
        break

    else:
        print('You have to enter a valid number')
````

````python
# and, in, not in 예시

a = input('input a : ')

l = ['apple', 'banana', 'grape', 'lemon', 'orange']
l2 = {'apple' : 'red', 'banana' : 'yellow', 'grape' : 'purple', 'lemon' : 'yellow', 'orange' : 'orange'}


if a in l and a in l2 :
    print(f'a is fruit! and color is {l2.get(a)}')

elif a not in l and a in list(l2.values()) :
    print('a is color')

else :
    print('I don\'t know what it is')
````
